```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings("ignore")

import os
pth=os.getcwd()
print(pth)

```

```python
for file in os.listdir(pth):
    if file.endswith(".csv"):
        print(file)
    
```

```python
drug_df=pd.read_csv("drug_approvals.csv",encoding='utf-8')
print("drug approval data loaded & number of records:",drug_df.shape[0])

#convert approval_date to pure date time
drug_df['approval_date']=pd.to_datetime(drug_df['approval_date'],errors='coerce')
```

```python
dis_df=pd.read_csv("disease_burden.csv",encoding='ascii')
print("diseases burden data loaded & number of records:",dis_df.shape[0])
```

```python
phrm_df=pd.read_csv("pharma_companies_financials.csv",encoding='ascii')
print("pharma company data loaded & number of records:",phrm_df.shape[0])
```

```python
bio_df=pd.read_csv("biotech_funding.csv",encoding='ascii')
print("biotech funding data loaded & number of records:",bio_df.shape[0])

##convert date column to pure date
bio_df['date']=pd.to_datetime(bio_df['date'],errors='coerce')
```

```python
clinic_df=pd.read_csv("clinical_trials.csv",encoding='ascii')
print("clinical data loaded & number of records:",clinic_df.shape[0])

##convert completion date column to pure date
clinic_df['completion_date']=pd.to_datetime(clinic_df['completion_date'],errors='coerce')
```
